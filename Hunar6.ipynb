```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = np.array([1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980])
heights = np.array([169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168,169.4, 170.9, 171, 173.9, 174.9, 176, 176.9, 177.1, 176.8])
```

```python
mean_height = np.mean(heights)
print(f"Mean Height: {mean_height:.2f} cm")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, heights, marker='o', label='Height over Years')
plt.axhline(y=mean_height, color='r', linestyle='--', label=f'Mean Height: {mean_height:.2f} cm')
plt.title('Height over the Years')
plt.xlabel('Year')
plt.ylabel('Height (Cm)')
plt.legend()
plt.grid(True)
plt.show()
```
